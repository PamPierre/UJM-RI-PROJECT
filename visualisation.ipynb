```python
import matplotlib.pyplot as plt
years = [1950, 1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3]

## Pour avoir une courbe avec le plot
plt.plot(years, gdp, color='red',marker='o',linestyle='solid')

#Titre
plt.title("Valeur de GPD")

# label
plt.ylabel("Milliards de dollards")
plt.xlabel("Année par 10")
plt.show()
```

```python
# diagramme en batton
plt.bar(years,gdp)
#Titre
plt.title("Valeur de GPD")

# label
plt.ylabel("Milliards de dollards")
plt.xlabel("Année par 10")
plt.show()
```

```python
#For oscars
movies = ['AH','BH', 'Casa', 'Ganh','West']
nb_oscar = [5,11,3,8,10]

xs = [i+0.1 for i,_ in enumerate(movies)]
plt.bar(xs, nb_oscar)
plt.ylabel("Nombre d'oscar")
plt.title("Mes films préférés")

#"Labelisé" l'axe des x avec le nom des films centrés sur le bâton
plt.xticks(xs, movies)
plt.show()
```

```python
variance = [pow(2,i) for i in range(9)] ##[1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [i for i in reversed(variance)] ## [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x,y in zip(variance, bias_squared)]

xs =list(range(0,len(variance), 1))# [i for i,_ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label="Variance")
plt.plot(xs, bias_squared, 'r-', label="Bias^2")
plt.plot(xs, total_error, 'b:', label="Eurreur total")

plt.legend(loc=9)
plt.xlabel("Complexité du model")
plt.title("Compromis entre biais et variance")
plt.show()
```

```python
friends = [70,60,72,63,71,64,60,64,67]
minutes=[175,170,205,120,220,130,150,145,190]
labels = ['a','b','c', 'd','e','f','g','h','i']
plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, xy=(friend_count, minute_count), xytext=(0, 4), textcoords='offset points')

plt.xlabel("Nombre d'amis")
plt.ylabel("Nombre de minutes passé par jour")
plt.title("Minutes par jour vs nombre d'amis")
plt.show()
```
